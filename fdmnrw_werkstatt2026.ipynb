## Calculate the Checksum
A checksum is a short fingerprint of a file's content. 
It is produced by a hash function (like MD5, SHA256, etc.).

## Why we need checksums

Checksums are mainly used for data integrity verification.
Example workflow:

- Compute checksum before upload
- Upload file
- Compute checksum after download
- Compare

If checksums match:
- file was transferred correctly

If they differ:
- file was corrupted during transfer

This is very important for large scientific datasets.
